```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

## Генеруємо датасет

```python
X = np.linspace(0, 1, 100)
```

```python
noise = np.random.normal(0, 0.1, 100)  # Генеруємо шум
Y = 2 * X + 1 + noise
```

```python
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Згенерований датасет')
plt.show()
```

## Робимо лінійну регресію

```python
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),  # Один вхідний шар з однією ознакою
    tf.keras.layers.Dense(units=1)  # Один вихідний шар (одна вага та одне зміщення)
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
model.fit(X, Y, epochs=100)
```

```python
Y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, label='Справжні дані')
plt.plot(X, Y_pred, color='red', linewidth=3, label='Прогноз моделі')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Лінійна регресія за допомогою TensorFlow')
plt.legend()
plt.show()
```
